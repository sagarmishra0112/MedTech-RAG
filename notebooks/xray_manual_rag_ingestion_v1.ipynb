```python
# text extraction using llamaparse
from llama_cloud_services import LlamaParse

parser = LlamaParse(
  api_key= os.environ.get("llamacloud_key"),
  
  tier="agentic_plus",
  version="latest",
  high_res_ocr=True,
  adaptive_long_table=True,
  outlined_table_extraction=True,

  # Whether to output tables as HTML in the markdown output
  output_tables_as_HTML=True,

  # Whether to use precise bounding box extraction (experimental)
  precise_bounding_box=True,

  # The page separator
  page_separator="\n\n---\n\n",

  # The maximum number of pages to parse
  max_pages=0,
)

# sync
result = parser.parse("D:\FinanceRAG\data\Allengers 100.pdf")

# get the llama-index markdown documents
markdown_documents = result.get_markdown_documents(split_by_page=True)


full_text =""
for page in result.pages:
    full_text += page.md + "\n\n"
    print(page.md)
```
